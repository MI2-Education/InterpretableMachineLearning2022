# Local Explanations of Breast Cancer Severity Models using Ceteris Paribus

The UCI ML Breast Cancer Wisconsin dataset, provided in sci-kit learn package, was created based on results of FNA procedure which 569 patients underwent to asses the severity of their breast cancer. Because the FNA procedure results in an image, the experts encoded 30 variables out of it to create tabular data. Detailed information, including simple exploratory analysis and modelling, can be found in [previous report on this topic](https://github.com/MI2-Education/InterpretableMachineLearning2022/blob/main/Homeworks/HW1/ZuzannaKwiatkowska/zuzanna_kwiatkowska_report.ipynb) or [Sci-Kit Learn Documentation](https://scikit-learn.org/stable/datasets/toy_dataset.html#breast-cancer-dataset).

The target variable in this dataset is binary, meaning it encodes:
- patients with malignant tumor as 0,
- patients with benign tumor as 1,

and there are no other classes of interest present in this task.

The aim of this report is to:
- create 2 simple models for solving the task (including 1 tree based),
- explain the predictions of the model locally using Ceteris Paribus. 


## 1. Stability of CP profiles within the same model

It was very hard to find examples which would show different feature dynamics for 2 observations. My hypothesis for that is that the model works almost perfectly, so the dynamics might not change a lot. The most interesting pair I found was the behaviours of area error for patient 411 (correctly predicted malignant tumor) and 237 (correctly predicted benign tumor). In the former case, area error has a more rapid dynamic (for smaller values there's a high chance of tumor being predicted as malignant which decreases when area error grows), whereas for benign patient the feature has almost no dynamic at all.

## Closing remarks

Ceteris Paribus method is a good way to measure model response with respect to a single variable. One of the use case I see for such analysis is identification of redundant variables - if the variable has almost no dynamics and clearly suggests one class, it may be crucial for classification purposes. However, the features where we can observe a substantial dynamics may be important in recognizing the interpretable "if-else" rules which we can use to describe our models and data.
